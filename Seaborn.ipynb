```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.DataFrame({'No':[1,2,3,4,5,6],
                   'Country':['Ind','Uk','Usa','Pak','Aig','Ban'],
                   'InflationRate':[8,9,10,11,10,9],
                   'Gdp' : [5,4,3,2,8,9],
                   'Sales' : [1000,4000,6000,8000,9000,6000]})
df
```

```python
import seaborn as sns
```

```python
# Line Plot
sns.lineplot(data = df,x=df['Country'],y=df['Sales'],linestyle='--',marker='o',color='green')
plt.show()
```

```python
# Bar Plot
sns.barplot(data = df,x=df['Country'],y=df['Sales'],color='green')
plt.show()
```

```python
# Count Plot
sns.countplot(data=df, x = df['Country'])
plt.show()
```

```python
# DisPlot
sns.displot(x = df['Sales'],kde = True)
plt.show()
```

```python
# Scatter Plot

sns.scatterplot(data=df,x=df['Gdp'],y=df['InflationRate'],color='red')
plt.show()
```

```python
# Scatter Plot

sns.scatterplot(data=df,x=df['Gdp'],y=df['InflationRate'],color='red',hue = df['Sales'])
plt.show()
```

```python
# Scatter Plot

sns.scatterplot(data=df,x=df['Gdp'],y=df['InflationRate'],color='red',hue = df['Country'])
plt.show()
```

```python
# RegPlot : Regression plot : Two Numwrical Column

sns.regplot(data=df,x=df['Gdp'],y=df['Sales'],ci=0.9,color='purple',marker='o',label='Regression line')
plt.legend()
plt.grid()
plt.ylim(500,10000)
plt.xlim(0,10)
plt.show()
```

```python
# Pair Plot
sns.pairplot(data=df)
```

```python
# joinPlot
sns.jointplot(data=df,x=df['Gdp'],y=df['Sales'],hue=df['InflationRate'])
plt.axhline(7000,linestyle = '--',color = 'lightblue')
plt.axvline(6.5,linestyle = '--',color = 'lightblue')
plt.show()
```
